## Uma loja de varejo deseja criar algumas métricas para acompanhar seus resultados. Para essa atividade, possuem as duas bases: varejo e cliente_varejo.


## Sua missão é fazer todo o tratamento e a união das bases para construção dessas métricas:

*   Departamento com vendas
*   Média de preço com frete por Nome de departamento
*   Quantidade de vendas por mês
*   Média de renda para cada tipo de canal de venda
*   Média de idade de clientes por bandeira



## Importando as bibliotecas a serem utilizadas

```python
import pandas as pd
import seaborn as sns
```

### Lendo o dataset

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

## 1) Analisando as categorias dos Dados e alterando algumas informações

```python
vendas.dtypes
```

```python
vendas.info( )
```

### Verificando com NUNIQUE

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

### Renomeando as colunas do DATASET

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

### Retirando os espaçoes em Branco e substituindo pelu underscore

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

### Tratando dados NULOS

```python
# Verificando os dados NULOS
vendas[vendas["estado"].isnull()]
```

### Substituindo valores NULOS

```python
vendas["estado"].fillna("MS", inplace=True) # a mesma coisa, mas escrevendo diferente ==> vendas["estado"] = vendas["estado"].fillna("MS")
vendas.info()
```

### Substituindo valores NULOS, pela média de todos os valores

```python
media_preco = vendas["Preço"].mean()
```

```python
vendas["Preço"].fillna(media_preco, inplace=True)
vendas.info()
```

##### **OBS: Existem algumas linhas na coluna "Preço" com valor errado, pois o valor do preço tem que ser menor que o valor da venda+frete na coluna "Preço_com_frete"**

```python
preco_errado = vendas.query("Preço > Preço_com_frete")

vendas_preco_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_preco_correto.describe()
```

# Pergunta 1 : Departamento com mais vendas




```python
vendas_preco_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

# Pergunta 2 : Média de preço com frete por nome de departamento

```python
round(vendas_preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(), 2) # round 2 para 2 casas decimais
```

# Pergunta 3 : Quantidade de vendas por mês

```python
# Criando a coluna com o mês
vendas_preco_correto["mes"] = vendas_preco_correto["Data"].dt.month
vendas_preco_correto.head()
```

```python
vendas_preco_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

# UNINDO AS DUAS TABELAS COM JOIN


```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
cliente[["idade", "renda"]].describe()
```

## Transfrmando a renda para o tipo de dados FLOAT


```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

## Como há a coluna "cliente_log" nos 2 datasets, dá para uní-las com o JOIN


```python
vendas_cliente = vendas_preco_correto.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

# Pergunta 4 : Média de renda para cada tipo de canal de venda

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(), 2)
```

```python
agg_idcanal_renda
```

# Pergunta 5: Média de idade de clientes por bandeira

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(), 2)
```

```python
agg_idade_bandeira
```

# APRESENTANDO OS DADOS COM DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
```

## Definindo o tamanho da figura

```python
fig_size = plt.figure(figsize=(10,8))
```

```python
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")

plt.ylabel("Renda Média")

plt.title("Renda média por canal de venda", loc="center", fontsize=18, fontweight="bold", color="black")
```
